load model

```python
from transformers import pipeline

generator = pipeline("text-generation", model="microsoft/DialoGPT-medium")
```

email generator

```python
def email_reply(text):
    prompt = f"Reply professionally to this email:\n{text}"
    output = generator(prompt, max_length=150, do_sample=True)
    return output[0]['generated_text']
```

cover letter generator


```python
def cover_letter(resume, job):
    prompt = f"""
Resume: {resume}
Job: {job}

Write a professional cover letter.
"""
    output = generator(prompt, max_length=250)
    return output[0]['generated_text']
```

```python
def travel_plan(place, days, interests):
    prompt = f"""
Plan a {days}-day trip to {place}.
Interests: {interests}.
Give itinerary.
"""
    output = generator(prompt, max_length=300)
    return output[0]['generated_text']
```

```python
def chatbot(msg):
    output = generator(msg, max_length=120)
    return output[0]['generated_text']
```

```python
import gradio as gr

with gr.Blocks() as app:

    gr.Markdown("# 🚀 FREE AI Business Automation Suite")

    with gr.Tab("Email Reply"):
        inp = gr.Textbox()
        out = gr.Textbox()
        gr.Button("Generate").click(email_reply, inp, out)

    with gr.Tab("Cover Letter"):
        r = gr.Textbox()
        j = gr.Textbox()
        out = gr.Textbox()
        gr.Button("Generate").click(cover_letter, [r, j], out)

    with gr.Tab("Travel Planner"):
        p = gr.Textbox()
        d = gr.Number()
        i = gr.Textbox()
        out = gr.Textbox()
        gr.Button("Plan").click(travel_plan, [p, d, i], out)

    with gr.Tab("Chatbot"):
        m = gr.Textbox()
        out = gr.Textbox()
        gr.Button("Ask").click(chatbot, m, out)

app.launch(share=True, inline=False)
```
